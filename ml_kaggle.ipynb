```python
import pandas as pd
melb_data = pd.read_csv('/content/sample_data/melb_data.csv')

# print the columns of the data
melb_data.columns
```

```python
melb_data = melb_data.dropna(axis=0)
melb_data
```

```python
# Target feature
y = melb_data.Price
y.head()
```

```python
# select features
melb_features = ['Rooms', 'Bathroom', 'Landsize', 'Lattitude', 'Longtitude']
X = melb_data[melb_features]

X
```

```python
# show some relations with the 
X.describe()
```

```python
X.head()
```

```python
# building the model
from sklearn.tree import DecisionTreeRegressor

melb_model = DecisionTreeRegressor(random_state=1)


# fit model
melb_model.fit(X,y)
```

```python
# making predictions
print("Making predictions for the first five house")
print(X.head())
print("\n the predictions are ")
print(melb_model.predict(X.head()))
```
