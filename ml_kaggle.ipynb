```python
import ipywidgets as widgets
from IPython.display import display
import os

uploader = widgets.FileUpload(
    accept='.csv',
    multiple=False,
    description='Select CSV from Downloads'
)
display(uploader)
```

```python
import os
import shutil

# Create your desired folder (e.g., 'my_dataset_folder')
folder_name = 'my_dataset_folder'
os.makedirs(folder_name, exist_ok=True)  # creates if not exists

# Move the uploaded file into the folder
if uploader.value:
    uploaded_filename = next(iter(uploader.value))  # e.g. 'my_data.csv'
    source_path = uploaded_filename  # it's already in /content/
    dest_path = os.path.join(folder_name, uploaded_filename)

    shutil.move(source_path, dest_path)
    print(f"File moved to: /{folder_name}/{uploaded_filename}")
else:
    print("Upload a file first using the button above.")
```
