# creating lists from values

```python
import pandas as pd

last_names = ['Annorbah', 'Tetteh', 'Cudjoe']
first_names = ['Michael', 'Jefferey', 'Nii']

df = pd.DataFrame({
    'first_name': first_names,
    'last_name': last_names
})
df
```

# renaming multiple columns

```python
import pandas as pd

df = pd.DataFrame({
    'Year':  [2020, 2021, 2022, 2023, 2024],
    'Top Animal': ['Cat', 'Dog', 'Hen', 'Fish','Rat']
})

df.rename(columns={
    'Year': 'Calender Year',
    'Top Animal': 'Animal'
}, inplace=True)

df

```

# query by variable name

```python
import pandas as pd

last_names = ['Annorbah', 'Tetteh', 'Cudjoe']
first_names = ['Michael', 'Jefferey', 'Nii']

df = pd.DataFrame({
    'first_name': first_names,
    'last_name': last_names
})

foo = "Annorbah"

df.query('last_name == @foo')
```
